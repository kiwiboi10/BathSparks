```python
import pandas as pd
from PIL import Image, ImageDraw, ImageFont, ImageFilter
import os
import textwrap

df = pd.read_excel(r'Path of downloaded excel file from microsoft submission form')
```

```python
from PIL import ImageDraw

def wrap_text(text, font, max_width):
    """Wrap text to ensure it doesn't exceed max width."""
    lines = text.splitlines()
    wrapped_text = []

    draw = ImageDraw.Draw(Image.new("RGB", (1, 1)))  # Dummy image for text measurements

    for line in lines:
        words = line.split()
        current_line = ""
        for word in words:
            test_line = f"{current_line} {word}".strip()
            text_width = draw.textbbox((0, 0), test_line, font=font)[2]
            if text_width <= max_width:
                current_line = test_line
            else:
                if current_line:
                    wrapped_text.append(current_line)
                current_line = word  # Start a new line with the word
        if current_line:
            wrapped_text.append(current_line)

    return '\n'.join(wrapped_text)
```

```python
# Filter by date
folder_name = r'Folder path where excel sheet is stored'
date_to_filter = 'todays date'
filtered_df = df
```

```python
for index, row in filtered_df.iterrows():
        # Create an image with the optimal aspect ratio for Instagram posts (4:5)
        img = Image.new('RGB', (1080, 1080), color='black')
        d = ImageDraw.Draw(img)

        # Load a font
        font = ImageFont.truetype('arial.ttf', 45)
        small_font = ImageFont.truetype('arial.ttf', 45)

        # Wrap the message text
        wrapped_message = wrap_text(row['Send a message to the Bath Community!!!'], font, img.width - 150) # Padding of 80 on both sides
        message_lines = wrapped_message.splitlines()

        # Calculate total height of text + date + ID
        total_text_height = sum([font.getsize(line)[1] for line in message_lines]) + font.getsize("X")[1]*2  # 2 line breaks for date and ID
        text_y = (img.height - total_text_height) // 2

        # Print the main message
        for line in message_lines:
            text_width, text_height = d.textsize(line, font=font)
            d.text((80, text_y), line, fill='white', font=font)  # Start text 80 pixels from the left
            text_y += text_height

        # Add a line break before date and ID
        text_y += font.getsize("X")[1]

        # Print date on the left side
        date_str = f"{row['Start time'].strftime('%d/%m/%Y')}"
        d.text((80, text_y), date_str, fill='white', font=small_font)

        # Print ID on the right side
        id_str = f"{row['ID']}"
        id_width, _ = d.textsize(id_str, font=small_font)
        d.text((img.width - id_width - 80, text_y), id_str, fill='white', font=small_font) # Padding of 80 pixels from the right

        # Save the image
        img.save(os.path.join(folder_name, f"{row['ID']}.png"))

print(f"Images generated in folder: {folder_name}")
```
